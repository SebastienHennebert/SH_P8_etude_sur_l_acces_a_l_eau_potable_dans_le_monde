```python
total_pop = pop[pop['Granularity'] == 'Total'][['Country', 'Year', 'Population']]

total_pop
```

# VUE NATIONALE

```python
# Sélection de la granularité 'Rural' dans la table pop

Rural = pop.loc[pop['Granularity'] == 'Rural']
Rural = Rural.rename(columns = {'% total' : '% pop rurale'})
Rural
```

```python
# Fusion des tables Rural et region_country afin d'obtenir les continents

vue_nationale = pd.merge(Rural, region_country, on = 'Country', how = 'outer')
vue_nationale ['Year'] = vue_nationale['Year'].astype('Int64')
vue_nationale
```

```python
# Présence problématique de 'China, mainland' et 'China'

vue_nationale['Country'].unique()
```

```python
# Suppression des lignes 'China' puis correction du pays 'China, mainland' en 'China' afin de ne garder que les données
## de la Chine seule sans Macao, Hong-Kong et Taïwan

vue_nationale = vue_nationale.loc[vue_nationale['Country'] != 'China']
vue_nationale['Country'] = vue_nationale['Country'].replace('China, mainland', 'China')
vue_nationale
```

```python
# Vérification de la ligne de code précédente

vue_nationale['Country'].unique()
```

```python
# Vérification des intitulés des pays, particulièrement pour la Chine
stabilite_pol['Country'].unique()
```

```python
# Modification de 'China, mainland' en 'China'

stabilite_pol['Country'] = stabilite_pol['Country'].replace('China, mainland', 'China')
stabilite_pol['Country'].unique()
```

```python
# Fusion des table vue_nationale et stabilité politique. Modification au préalable de la colonne 'year' en 'Int64' vs 'int64'

stabilite_pol ['Year'] = stabilite_pol['Year'].astype('Int64')
vue_nationale = pd.merge(vue_nationale, stabilite_pol, on = ['Year', 'Country'], how = 'outer')
vue_nationale = vue_nationale.drop(columns = {'Granularity_x', 'Granularity_y'})
vue_nationale = vue_nationale.rename(columns = {'Political_Stability' : 'indice stabilité politique'})
vue_nationale
```

```python
# Sélection de la granularité Total dans mortalite_eau et suppression de la colonne 'Granularity'

mortalite = mortalite_eau.loc[mortalite_eau['Granularity'] == 'Total']
mortalite = mortalite.drop(columns = {'Granularity', 'Year'})
mortalite
```

```python
# Fusion des tables mortalite et vue_nationale

vue_nationale = pd.merge(vue_nationale, mortalite, on = 'Country', how = 'outer')
vue_nationale
```

```python
# Export CSV du dataframe vue_nationale

vue_nationale.to_csv('vue_nationale.csv', index = False)
```

# INDICATEUR 1 : Taux d'accès à l'eau potable et population urbaine

```python
# Sélection de la granularité 'Urban' dans la table pop

Urban = pop.loc[pop['Granularity'] == 'Urban']
Urban = Urban.rename(columns = {'% total' : '% pop urbaine'})
Urban
```

```python
# Ajout de la table Eau au tableau final

Indicateur1 = pd.merge(Indicateur1, Eau, on = ['Country', 'Year'], how = 'outer')
Indicateur1
```

```python
# Suppression des colonnes de granularité non nécessaires

Indicateur1 = Indicateur1.drop(columns = ['Granularity_x', 'Granularity_y'])
Indicateur1
```

```python
# Export CSV du dataframe Indicateur1

Indicateur1.to_csv('Indicateur1.csv', index = False)
```

# Focus République Démocratique du Congo

```python
# Dans la table pop, sélection des éléments sur la RDC sur la granularité 'Total', 'Urban' et 'Rural'

focusRDC = pop.loc[(pop['Country'] == 'Democratic Republic of the Congo') & (pop['Granularity'].isin(['Total', 'Rural', 'Urban']))]
focusRDC
```

```python
# Dans la table acces_eau, sélection des éléments sur la RDC

eauRDC = acces_eau.loc[(acces_eau['Country'] == 'Democratic Republic of the Congo')]
eauRDC
```

```python
# Fusion des 2 tables

focusRDC = pd.merge(focusRDC, eauRDC, on  = ['Year', 'Granularity'], how = 'outer')
focusRDC = focusRDC.drop(columns = 'Country_y')
focusRDC = focusRDC.rename(columns = {'Country_x' : 'Country', '% total' : '% total pop'})
focusRDC
```

```python
# Dans la table stabilite_pol, sélection des éléments sur la RDC

stabilitepolRDC = stabilite_pol.loc[stabilite_pol['Country'] == 'Democratic Republic of the Congo']
stabilitepolRDC = stabilitepolRDC.drop(columns = 'Country')
stabilitepolRDC
```

```python
# Fusion des tables

focusRDC = pd.merge(focusRDC, stabilitepolRDC, on = ['Year', 'Granularity'], how = 'outer')
focusRDC
```

```python
# Dans la table mortalite_eau, sélection des éléments sur la RDC

mortaliteRDC = mortalite_eau.loc[(mortalite_eau['Country'] == 'Democratic Republic of the Congo') & (mortalite_eau['Granularity'] == 'Total')]
mortaliteRDC = mortaliteRDC.drop(columns = ['Year', 'Country'])
mortaliteRDC
```

```python
# Fusion pour obtenir la table finale

focusRDC = pd.merge(focusRDC, mortaliteRDC, on = 'Granularity', how = 'left')
focusRDC
```

```python
# Export CSV du dataframe focusRDC

focusRDC.to_csv('focusRDC.csv', index = False)
```
